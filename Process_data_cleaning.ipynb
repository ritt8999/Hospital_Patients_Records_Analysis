```python
df['Name'] = df['Name'].str.title()
df['Doctor'] = df['Doctor'].str.title()
df['Hospital'] = df['Hospital'].str.title()
df['Insurance Provider'] = df['Insurance Provider'].str.title()
df['Medication'] = df['Medication'].str.title()
df['Medical Condition'] = df['Medical Condition'].str.title()
```

```python
df['Gender'] = df['Gender'].str.capitalize()
```

```python
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

```python
df[['Name','Gender','Medical Condition','Doctor','Hospital']].head(10)
```

```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
```

```python
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'])
```

```python
df['Length_of_Stay'] = (df['Discharge Date'] - df['Date of Admission']).dt.days
```

```python
df[['Date of Admission','Discharge Date','Length_of_Stay']].head(10)
```

```python
df[['Date of Admission','Discharge Date','Length_of_Stay']].head(10)
```

```python
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[0, 18, 35, 50, 70, 150],
    labels=['0-18', '19-35', '36-50', '51-70', '70+']
)
```

```python
df['Billing_Category'] = pd.cut(
    df['Billing Amount'],
    bins=[0, 10000, 30000, df['Billing Amount'].max()],
    labels=['Low', 'Medium', 'High']
)
```

```python
df[['Age', 'Age_Group', 'Billing Amount', 'Billing_Category']].head(15)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['Age'], kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
sns.histplot(df['Billing Amount'], kde=True)
plt.title("Billing Amount Distribution")
plt.show()
```

```python
sns.countplot(data=df, x='Medical Condition')
plt.xticks(rotation=45)
plt.title("Medical Conditions Count")
plt.show()
```

```python
sns.countplot(data=df, x='Gender')
plt.title("Gender Distribution")
plt.show()
```

```python
sns.countplot(data=df, x='Age_Group', hue='Medical Condition')
plt.title("Age Group vs Medical Condition")
plt.xticks(rotation=45)
plt.show()
```

```python
sns.countplot(data=df, x='Billing_Category', hue='Admission Type')
plt.title("Billing Category vs Admission Type")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.select_dtypes(include=['int64','float64']).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
df['Admission_Weekday'] = df['Date of Admission'].dt.day_name()
```

```python
df['Stay_Category'] = pd.cut(
    df['Length_of_Stay'],
    bins=[0, 3, 10, df['Length_of_Stay'].max()],
    labels=['Short Stay', 'Medium Stay', 'Long Stay']
)
```

```python
df['Admission_Month'] = df['Date of Admission'].dt.month_name()
df['Admission_Year'] = df['Date of Admission'].dt.year
```
